# **Importing libraries**

```python
import os
from getpass import getpass
```

```python
import os.path
from llama_index.core import (
    # It allows you to create an index from a set of documents,
    # where the documents are represented as vectors (embeddings) in a high-dimensional space.
    VectorStoreIndex,
    # This class is a reader that reads files from a specified directory.
    # It can be used to load documents from a local directory into LlamaIndex
    SimpleDirectoryReader,
    # This class provides a context manager for storing and loading indexes
    StorageContext,
    #  This function is used to load an existing index from a storage backend.
    load_index_from_storage,
)
```

# **loading model**

```python
# Get pass is helpful to set keys into environments without exposing them
google_api_key = getpass()
os.environ['GOOGLE_API_KEY'] = google_api_key
```

```python
# to list available Gen AI models
import google.generativeai as genai

for m in genai.list_models():
    if "generateContent" in m.supported_generation_methods:
        print(m.name)
```

```python
# Using the model
from llama_index.llms.gemini import Gemini

llm = Gemini(model="models/gemini-pro")
```

# **Embedding/Indexing**

```python
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.core import Settings

# loading a small embedding model from hugging face
Settings.embed_model = HuggingFaceEmbedding(
    model_name="BAAI/bge-small-en-v1.5"
)
```

```python
PERSIST_DIR = "/content/drive/MyDrive/rag/storage" # replace this folder structure with a folder structure you like
# The 'storage' folder is the persistent storage directory where the index will be stored or loaded from
# 'Rag' folder has all the required stories to perform tasks on.

# check if storage already exists
if not os.path.exists(PERSIST_DIR):
    # load the documents and create the index
    documents = SimpleDirectoryReader("/content/drive/MyDrive/rag").load_data() # replace this folder structure with a folder structure you like
    index = VectorStoreIndex.from_documents(documents)
    # store it for later
    index.storage_context.persist(persist_dir=PERSIST_DIR)
else:
    # load the existing index
    storage_context = StorageContext.from_defaults(persist_dir=PERSIST_DIR)
    index = load_index_from_storage(storage_context)

documents = SimpleDirectoryReader("/content/drive/MyDrive/rag").load_data() # replace this folder structure with a folder structure you like
```

```python
print(documents)
```

# **Summerization**

```python
# we can now query the index
query_engine = index.as_query_engine(llm=llm)
response = query_engine.query("Summarize Alexander story in 300 words")
print(response)
```
